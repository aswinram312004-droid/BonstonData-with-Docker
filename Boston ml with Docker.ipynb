<div style="font-family: Arial; line-height:1.7;">

  <h2 style="color:#1f4bd8;"> STANDARDIZATION NA ENNA?</h2>

    <tr style="background:#e8edff;">
      <th style="border:1px solid #ccc; padding:8px;">Feature</th>

  <br>

  <h3 style="color:#1f4bd8;"> ML PROBLEM WITHOUT SCALING</h3>

  <ul>
    <li>TAX values romba <b>perusa</b> irukum</li>
    <li>CRIM values romba <b>chinna</b> irukum</li>
  </ul>

  <p> Model think pannum:</p>

  <div style="background:#fff7e6; padding:12px; border-left:6px solid #ff9800; border-radius:6px;">
    <p><i>TAX romba important</i> </p>
    <p>(Just because numbers big  not because actual importance)</p>
  </div>

  <br>

  <h3 style="color:#1f4bd8;"> SOLUTION: STANDARDIZATION</h3>

  <p><b>Formula:</b></p>

  <pre style="background:#0f172a; color:#e5e7eb; padding:12px; border-radius:6px;">
z = (x - mean) / std
  </pre>

  <ul>
    <li> Mean becomes <b>0</b></li>
    <li> Standard Deviation becomes <b>1</b></li>
  </ul>

  <br>

  <h3 style="color:#1f4bd8;"> WHAT STANDARDIZATION DOES?</h3>

  <table style="border-collapse:collapse; width:70%;">
    <tr style="background:#e8edff;">
      <th style="border:1px solid #ccc; padding:8px;">Feature</th>
      <th style="border:1px solid #ccc; padding:8px;">Before</th>
      <th style="border:1px solid #ccc; padding:8px;">After Scaling</th>
    </tr>
    <tr>
      <td style="border:1px solid #ccc; padding:8px;">CRIM</td>
      <td style="border:1px solid #ccc; padding:8px;">0.006</td>
      <td style="border:1px solid #ccc; padding:8px;">-0.7</td>

  </table>

  <p> Now all features are on <b>equal footing</b> </p>

  <br>

  <h3 style="color:#1f4bd8;"> STEP-BY-STEP CODE (VERY IMPORTANT)</h3>

  <pre style="background:#0f172a; color:#e5e7eb; padding:14px; border-radius:6px;">
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(X_test)
  </pre>

  <br>

  <h3 style="color:#d32f2f;"> WHY <code>fit_transform()</code> ON TRAIN ONLY?</h3>
  <p><b>MOST IMPORTANT CONCEPT </b></p>

  <div style="background:#fff4f4; padding:14px; border-left:6px solid #e53935; border-radius:6px;">
    <ul>
      <li> Test data-la <b>fit</b> panna koodathu</li>
      <li> <b>Data Leakage</b> nadakkum</li>
    </ul>
  </div>

  <br>

  <div style="background:#e8f5e9; padding:14px; border-left:6px solid #4caf50; border-radius:6px;">
    <p><b>Correct Logic:</b></p>
    <ul>
      <li>Train data  <b>model learning</b></li>
      <li>Test data  <b>unseen data</b></li>
    </ul>
    <p>
       Test data information <b>train-ku leak aaga koodathu</b>
    </p>
  </div>

  <br>

  <div style="background:#f3f6ff; padding:14px; border-left:6px solid #1f4bd8; border-radius:6px;">
    <p><b> FINAL INTERVIEW TAKEAWAY</b></p>
    <p>
      <i>Standardization prevents models from giving unfair importance to large-scale features and avoids data leakage when applied correctly.</i>
    </p>
  </div>

## metrics Concept 

### R2_score to check the relationship between the y_test and  regression prediction

```python
from sklearn.metrics import r2_score as scr
score=scr(y_test,reg_predict)
print(score)
```

Ne model predict pannita 

<d> Ipa client/company  keakuranga prediction  ok but intha Model average-aa evlo mistake pannuthu?</d>

Model evlo thappu pannuthu?

 Athuku answer dhaan
MAE, MSE, RMSE

| Metric | Focus             |
| ------ | ----------------- |
| MAE    | Average mistake   |
| RMSE   | Big mistakes      |
| MSE    | Mathematical loss |


```python
from sklearn.metrics import mean_squared_error as ms
from sklearn.metrics import mean_absolute_error as ma

print("Mean Square error (mse)",ms(y_test,reg_predict))
print("mean Average Error (mae)",ma(y_test,reg_predict))
print("Root Mean Square Error (RMSE)",np.sqrt(ms(y_test,reg_predict)))
```
